## <font color = 'cornflowerblue' size=4>Introduction</font>

In this notebook we will look at student based features.

Caveat: we only have data that spans 2 weeks. So we will need to leave the beginning few days out (as they will need to be the basis for the features based on student behavior, so as to prevent data leakage.)

Also, some of these features depend on the student's choices for that day. If they choose to spend a session reviewing old words vs learning new ones. 


numseen_day	avgp_forgot_day	numwordspracticed_day	numcorrect_day	numquizzes_day	avgdelta_day	avgactivesessions	duration
Some of the aggregate features being considered:
1. Average Number of total questions per day
2. Average number of incorrect questions per day
3. Average number of words practiced per day
4. Average number of quizzes per day
5. Average delta per day (represents the ratio of old to new words they practice per day)
6. avgactivesesssions per day (how distracted they are)
7. duration: how long they study for per day

The features above depend on how many days students have logged in.

The dataset is filtered to students who have logged in at least x number of days, and aggregate features are calculated for 1 to x-1 days to see how well they separate performance of students.

Define function to:
1. draw box plots for a feature based on increasing number of days of history
2. get the dependence of difference in means of forgotten vs remembered on the number of days of history taken.

```python
def explore_agg_feature(y_,ylimits=None):

  plt.subplots(4,3,figsize=(20,20))
  plot_num = 1

  mean_nd0 = dict()
  mean_nd1 = dict()
  for nd in range(2,12):

    numberdays_history = nd

    #print(f"Number of students with a history of at least {numberdays_history} days logged in: {numdaysloggedin[numdaysloggedin>=numberdays_history].shape[0]}")

    # Users with x number of days logged in
    users_withxsessions = numdaysloggedin[numdaysloggedin==numberdays_history].index

    # aggregates for each date for users with x number of log in days
    Four = userdates.loc[userdates['user_id'].isin(users_withxsessions),:]

    # rank dates for each user
    Four['daterank'] = Four.groupby('user_id')['Date'].rank()

    # calculate interval between days
    Four['Prev_Date'] = Four.groupby('user_id')['Date'].shift(1)
    Four['Deltaday'] = Four['Date'] - Four['Prev_Date']

    # convert to integer
    Four['Deltaday'] = Four['Deltaday'].dt.days

    # get aggregate features for 1 to x-1 dats
    Fouraggs = Four.loc[Four['daterank']<numberdays_history].groupby('user_id')[['numseen_day','avgp_forgot_day','avg_history_seen','avg_history_frac','numwordspracticed_day','numcorrect_day','numquizzes_day','avgdelta_day','avgactivesessions','duration','Deltaday']].mean()
    

    # get xth date for each user 
    fourthrankdates = Four.loc[Four['daterank']==numberdays_history,['user_id','Date']]

    # couple user_id and the xth date
    fourthrankdates['user_date_tup'] = fourthrankdates[['user_id','Date']].apply(lambda x:frozenset([x[0],str(x[1])]),axis=1)

    # join the aggregate features to the fourth dates
    fourthrankdates_features = pd.merge(left = fourthrankdates, right = Fouraggs, left_on = 'user_id',right_index=True) 

    # user_date_tup = q2train[['user_id','Date']].apply(lambda x: frozenset([x[0],str(x[1])]), axis=1)
    # q2train['user_date_tup'] = user_date_tup

    # filter out only the records of xth date 
    fourthday = q2train.loc[q2train['user_date_tup'].isin(fourthrankdates_features['user_date_tup']),:]

    fourthday1 = pd.merge(left = fourthday,right =fourthrankdates_features, left_on = 'user_id',right_on= 'user_id')

    mean_nd0[nd] = fourthday1.loc[fourthday1['p_forgot_bin']==0,y_].median()
    mean_nd1[nd] = fourthday1.loc[fourthday1['p_forgot_bin']==1,y_].median()

    plt.subplot(4,3,plot_num)
    sns.boxplot(x = fourthday1['p_forgot_bin'],y = fourthday1[y_])
    if ylimits is not None:
      plt.ylim(ylimits)
    plt.title(f"mean {y_} for {numberdays_history-1} days history")
    plot_num +=1

  plt.show()

  plt.figure()
  plt.plot(pd.Series(mean_nd1).index,pd.Series(mean_nd1),'.-',label = "Median for forgotten",color='darkorange')
  plt.plot(pd.Series(mean_nd1).index,pd.Series(mean_nd0),'.-',label = "Median for remembered",color='cornflowerblue')
  plt.title(f"Medians for {y_} vs number of history days taken")
  plt.xlabel("Number of days in history")
  plt.ylabel(f"Median {y_}")
  plt.legend()
  plt.show()
```

### <font color = 'cornflowerblue' size=3>Does past performance for all words predict present day performance?</font>

Exploring avgp_forgot_day: The average fraction of forgotten words per day

The boxplots below plot the average performance on previous x days for forgotten and remembered records. 

The chart at the end plots the difference in means of avgp_forgot_day vs the number of days of history taken.

The separation gets better at higher days (11), but the caveat is much fewer students. 

```python
explore_agg_feature('avgp_forgot_day')
```

It seems that past performance does predict current performance consistently (refer boxplots) - if the fraction of forgotten words averaged over $x-1$ days is high, the probability of forgetting words on $x^{th}$ day is high. 

Counterituitvely, if you take more days of history to calculate aggregates of performance, the separation (difference in medians) actually reduces rather than increases. This is probably because with 1-2 days of history, the aggregates includes a lot of irregular students. With more days of history, only regular students are remaining (only 576 students have logged in for at least 11 days. ), so the two groups of forgotten vs remembered are not as separated, since regular students will learn better. 

It might be helpful to take 2-4 days history to separate out irregular from regular students.

For more regular students (more than 6 days of history), students who choose to study words practiced long ago tend to answer more questions incorrectly (they've forgotten those words), whereas for irregular students 2-4 days of history, students who remember words long ago tend to remember them.  

It might be helpful to use 3 days history of avgdelta_day to separate out irregular from regular students.

## <font color = 'cornflowerblue' size=4>Preprocessing student features</font>

Preprocessing for student-based features only with 3 days of history for the following features.

1. avg_history_frac
2. avgdelta_day
3. avgp_forgot_day
4. number of words practiced per day. 

The 3-day aggregates are computed and joined with the dataframe. 

Statistical tests for them are done once they are computed. 

```python
numberdays_history = 11

#Users with x number of days logged in
users_withxsessions = numdaysloggedin[numdaysloggedin==numberdays_history].index

# aggregates for each date for users with x number of log in days
Four = userdates.loc[userdates['user_id'].isin(users_withxsessions),:]

# rank dates for each user
Four['daterank'] = Four.groupby('user_id')['Date'].rank()

# calculate interval between days
Four['Prev_Date'] = Four.groupby('user_id')['Date'].shift(1)
Four['Deltaday'] = Four['Date'] - Four['Prev_Date']

# convert to integer
Four['Deltaday'] = Four['Deltaday'].dt.days

# get aggregate features for 1 to x-1 dats
Fouraggs = Four.loc[Four['daterank']<numberdays_history].groupby('user_id')[['avgp_forgot_day','avg_history_frac','numwordspracticed_day','avgdelta_day']].mean()


# get xth date for each user 
fourthrankdates = Four.loc[Four['daterank']==numberdays_history,['user_id','Date']]

# couple user_id and the xth date
fourthrankdates['user_date_tup'] = fourthrankdates[['user_id','Date']].apply(lambda x:frozenset([x[0],str(x[1])]),axis=1)

# join the aggregate features to the fourth dates
fourthrankdates_features = pd.merge(left = fourthrankdates, right = Fouraggs, left_on = 'user_id',right_index=True) 

# user_date_tup = q2train[['user_id','Date']].apply(lambda x: frozenset([x[0],str(x[1])]), axis=1)
# q2train['user_date_tup'] = user_date_tup

# filter out only the records of xth date 
fourthday = q2train.loc[q2train['user_date_tup'].isin(fourthrankdates_features['user_date_tup']),:]

# joining with main dataframe records based on user_id
q2trainX3d_sfonly = pd.merge(left = fourthday,right =fourthrankdates_features, left_on = 'user_id',right_on= 'user_id')

# dropping extraneous columns:
q2trainy3d_sfonly = q2trainX3d_sfonly['p_forgot_bin']
q2trainX3d_sfonly = q2trainX3d_sfonly[['delta','history_seen','history_frac','simoverdiff','avgp_forgot_day','avg_history_frac','numwordspracticed_day','avgdelta_day']]
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score,\
 roc_auc_score, plot_roc_curve,f1_score,\
  confusion_matrix, classification_report,\
  precision_recall_curve


# creating dictionaries to store metrics for different knn models
valid_acc = dict()
train_acc = dict()

valid_AUC = dict()
train_AUC = dict()

train_cm = dict()
valid_cm = dict()

train_f1 = dict()
valid_f1 = dict()

train_prc = dict()
valid_prc = dict()

train_proba1 = dict()
valid_proba1 = dict()


# create subplots
figs, axes = plt.subplots(1, 2, figsize=(20, 10))


depths = range(2,14)

for d in depths:
  print(f"Using max depth = {d}")
  DTCmodel = DecisionTreeClassifier(max_depth = d,class_weight='balanced') # intitalize a Decision tree classifier model with a new max depth
  DTCmodel.fit(Xq2_train3d,yq2_train3d) 

  # predicting on training and validation sets
  y_train_predicted = DTCmodel.predict(Xq2_train3d) 
  y_valid_predicted = DTCmodel.predict(Xq2_valid3d) 

  # soft predictions on training and validation data
  y_train_proba = DTCmodel.predict_proba(Xq2_train3d) 
  y_valid_proba = DTCmodel.predict_proba(Xq2_valid3d) 

  # saving metrics for training and validayion data
  # calculate accuracy
  valid_acc[d] = accuracy_score(yq2_valid3d, y_valid_predicted)
  train_acc[d] = accuracy_score(yq2_train3d,y_train_predicted)

  # getting roc_auc scores for training and validation sets 
  valid_AUC[d] = roc_auc_score(yq2_valid3d, y_valid_proba[:,1])
  train_AUC[d] = roc_auc_score(yq2_train3d, y_train_proba[:,1])

  # getting confusion matrices
  valid_cm[d] = confusion_matrix(yq2_valid3d, y_valid_predicted)
  train_cm[d] = confusion_matrix(yq2_train3d, y_train_predicted)

  # F1 -scores
  valid_f1[d] = f1_score(yq2_valid3d, y_valid_predicted)
  train_f1[d] = f1_score(yq2_train3d, y_train_predicted)

  # precision recall curves
  valid_prc[d] = precision_recall_curve(yq2_valid3d, y_valid_proba[:,1])
  train_prc[d] = precision_recall_curve(yq2_train3d, y_train_proba[:,1])

  # saving probabilities of positive class
  valid_proba1[d] = y_valid_proba[:,1]
  train_proba1[d] = y_train_proba[:,1]

  # plot ROC curve for training and validation sets
  plot_roc_curve(DTCmodel, Xq2_train3d, yq2_train3d, ax=axes[0],name=f"maxdepth = {d}")
  plot_roc_curve(DTCmodel, Xq2_valid3d, yq2_valid3d, ax=axes[1],name=f"maxdepth = {d}")

  # Add line y=x and titls
  for axis_num, plot_title in enumerate(['Training set', 'Validation set']):
    axes[axis_num].plot([0,1], [0,1], color='black', linestyle='--')
    axes[axis_num].set_title(plot_title)

plt.show()

```
